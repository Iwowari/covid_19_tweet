```python
from sklearn.model_selection import train_test_split
```

```python
train, valid = train_test_split(train_df, test_size= 0.2, stratify= train_df["target"])
```

```python
train.head()
```

```python
valid.head()
```

```python
train = train.drop("ID", axis =1)
train.columns= ["text", "labels"]

```

```python
valid = valid.drop("ID", axis =1)
valid.columns= ["text", "labels"]
```

```python
train
```

```python
model_args = ClassificationArgs(num_train_epochs=5, train_batch_size=32, overwrite_output_dir=True)
```

```python
%%capture
model = ClassificationModel(
    "roberta", "roberta-base", args=model_args
)
```

```python
model.train_model(train, eval_df=valid)
```

```python
result,model_outputs, wrong_predictions =model.eval_model(valid)
```
